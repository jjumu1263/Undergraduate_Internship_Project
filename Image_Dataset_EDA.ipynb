# **이미지 데이터에 대한 탐색적 데이터 분석(EDA, Exploratory Data Analysis)**

- 분석 목적·변수 확인 → 데이터 품질 점검 → 변수별 특성 확인 → 변수 간 관계 분석 등을 수행


## MNIST 데이터(숫자)

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets as ds
import seaborn as sns
import collections
```

```python
# 이미지 불러오기
(x_train,y_train),(x_test,y_test) = ds.mnist.load_data()
print("훈련 세트:", x_train.shape, y_train.shape) # 6000장, 28x28 크기, 흑백 이미지
print("테스트 세트:", x_test.shape, y_test.shape) # 1000장, 28x28 크기, 흑백 이미지

# cifar10 클래스명
class_names = [str(i) for i in range(10)]
```

```python
# 데이터 개수 세기
train_counter = collections.Counter(y_train.flatten())

print("\n[학습 데이터 클래스 분포]")
for idx, cnt in sorted(train_counter.items()):
    print(f"{class_names[idx]:>2}: {cnt}")
```

```python
# 테스트 데이터 클래스 분포 확인하기
# ---------Fill in the code----------
test_counter = collections.Counter(y_test.flatten())

print("\n[테스트 데이터 클래스 분포]")
for idx_test, cnt_test in sorted(test_counter.items()):
  print(f"{class_names [idx_test]:>2}: {cnt_test}")



# ------------------------------------
```

```python
# 학습 데이터 분포
train_counts = [train_counter[i] for i in range(10)]
plt.figure(figsize=(8,4))
plt.bar(class_names, train_counts, color='skyblue')
plt.title("MNIST Training Data Class Distribution", fontsize=16)
plt.xlabel("Class")
plt.ylabel("Number of Samples")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 테스트 데이터 분포
# graph option : color='salmon', title = "MNIST Test Data Class Distribution", fontsize=16

# ---------Fill in the code----------
test_counts = [test_counter[i] for i in range(10)]
plt.figure(figsize = (8,4))
plt.bar(class_names, test_counts, color = 'salmon')
plt.title("MNIST Test Data Class Distribution", fontsize = 16)
plt.xlabel("Class")
plt.ylabel("Number of Samples")
plt.grid(axis='y', linestyle = '--', alpha = 0.7)
plt.show()



# ------------------------------------
plt.show()
```

```python
# 10개 이미지 시각화
plt.figure(figsize=(24,3))
plt.suptitle('MNIST',fontsize=30)
for i in range(10):
  plt.subplot(1,10,i+1)
  plt.imshow(x_train[i],cmap='gray')
  plt.xticks([]); plt.yticks([])
  plt.title(str(y_train[i]),fontsize=20)
```

```python
# 첫 번째 이미지 픽셀값 확인(x_train안의 숫자를 변경하면 다른 이미지들도 확인 가능)
print(x_train[0])

# 첫번째 이미지의 최소, 최대 픽셀 값과 데이터 타입 확인
print("최소:", x_train[0].min(), "최대:", x_train[0].max())
print("dtype:", x_train[0].dtype)

# 첫 번째 이미지 시각화
plt.imshow(x_train[0], cmap='gray')
```

## CIFAR-10 데이터(클래스가 10개인 이미지 데이터셋)

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets as ds
import seaborn as sns
import collections
```

```python
# 이미지 불러오기
(x_train,y_train),(x_test,y_test) = ds.cifar10.load_data()
print("훈련 세트:", x_train.shape, y_train.shape) # 5000장, 32x32 크기, 3차원 컬러
print("테스트 세트:", x_test.shape, y_test.shape) # 1000장, 32x32 크기, 3차원 컬러

# cifar10 클래스명
class_names = ['airplane','car','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
# 데이터 개수 세기
train_counter = collections.Counter(y_train.flatten())
test_counter  = collections.Counter(y_test.flatten())

print("\n[학습 데이터 클래스 분포]")

# ---------Fill in the code----------
for idx1, cnt1 in sorted(train_counter.items()):
  print(f"{class_names[idx1]:>2} : {cnt1}")




# ------------------------------------

print("\n[테스트 데이터 클래스 분포]")

# ---------Fill in the code----------
for idx_test1, cnt_test1 in sorted(test_counter.items()):
  print(f"{class_names[idx_test1]:>2} : {cnt_test1}")




# ------------------------------------
```

```python
# 학습 데이터 분포
train_counts = [train_counter[i] for i in range(10)]
plt.figure(figsize=(8,4))
plt.bar(class_names, train_counts, color='skyblue')
plt.title("CIFAR-10 Training Data Class Distribution", fontsize=16)
plt.xlabel("Class")
plt.ylabel("Number of Samples")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 테스트 데이터 분포
# ---------Fill in the code----------
test_counts = [test_counter[1] for i in range(10)]
plt.figure(figsize = (8,4))
plt.bar(class_names, test_counts, color = 'salmon')
plt.title("CIFAR-10 Test Data Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Samples")
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)


# ------------------------------------
plt.show()
```

```python
# 10개 이미지 시각화
plt.figure(figsize=(24,3))
plt.suptitle('CIFAR-10',fontsize=30)

# ---------Fill in the code----------

for i in range(10) : #학습 데이터 이미지 시각화
  plt.subplot(1, 10, i + 1)
  plt.imshow(x_train[i])
  plt.xticks([]); plt.yticks([])
  plt.title(str(y_train[i]), fontsize = 20)


for i in range(10) : #테스트 데이터 이미지 시각화
  plt.subplot(1, 10, i + 1)
  plt.imshow(x_test[i])
  plt.xticks([]); plt.yticks([])
  plt.title(str(y_test[i]), fontsize = 20)
plt.show()
# ------------------------------------
```

```python
# 첫 번째 이미지 픽셀값(RGB) 출력
print(x_train[0])

# 첫번째 이미지의 최소, 최대 픽셀 값과 데이터 타입 확인
print("최소:", x_train[0].min(), "최대:", x_train[0].max())
print("dtype:", x_train[0].dtype)

# 첫 번째 이미지 시각화
ax1 = plt.subplot(2, 1, 1)
plt.imshow(x_train[0])
plt.ylabel('train')

ax2 = plt.subplot(2, 1, 2)
plt.imshow(x_test[0])
plt.ylabel('test')
```

## Fashion Mnist dataset EDA



```python
# 필요한 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets as ds
import seaborn as sns
import collections

```

```python
# 데이터 불러오기
(x_train, y_train), (x_test, y_test) = ds.fashion_mnist.load_data()
print("훈련 세트 :", x_train.shape, y_train.shape)
print("테스트 세트 :", x_test.shape, y_test.shape)

#클래스명
class_names = [str(i) for i in range(10)]
```

```python
# 데이터 개수 세기
train_counter = collections.Counter(y_train.flatten())
test_counter = collections.Counter(y_test.flatten())

print("\n[학습 데이터 클래스 분포]")

for idx, cnt in sorted(train_counter.items()) :
  print(f"{class_names[idx]:>2} : {cnt}")


print("\n[테스트 데이터 클래스 분포]")
for idx, cnt in sorted(test_counter.items()) :
  print(f"{class_names[idx]:>2} : {cnt}")
```

```python
# 데이터 분포 확인
train_counts = [train_counter[i] for i in range(10)]
plt.figure(figsize = (8,4))
plt.bar(class_names, train_counts, color = 'skyblue')
plt.title("Fashion MNIST Training Data Class Distribution", fontsize = 16)
plt.xlabel("Class")
plt.ylabel("Number of Samples")
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()

test_counts = [test_counter[i] for i in range(10)]
plt.figure(figsize = (8, 4))
plt.bar(class_names, test_counts, color = 'salmon')
plt.title("Fashion MNIST Test Data Class Distribution", fontsize = 16)
plt.xlabel("Class")
plt.ylabel("Number of Samples")
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()
```

```python
# 10개 이미지 시각화
#train
plt.figure(figsize = (24, 3))
plt.suptitle('Fashion MNIST Train', fontsize = 30)
for i in range(10) :
  plt.subplot(1, 10, i + 1)
  plt.imshow(x_train[i], cmap = 'gray')
  plt.xticks([]); plt.yticks([])
  plt.title(str(y_train[i]), fontsize = 20)

#test
plt.figure(figsize = (24, 3))
plt.suptitle('Fashion MNIST Test', fontsize = 30)
for i in range(10) :
  plt.subplot(1, 10, i + 1)
  plt.imshow(x_test[i], cmap = 'gray')
  plt.xticks([]); plt.yticks([])
  plt.title(str(y_test[i]), fontsize = 20)
```

```python
# 첫 번째 이미지 픽셀값(RGB) 출력
print(x_train[0]) #train
print(x_test[0]) #test


# 첫번째 이미지의 최소, 최대 픽셀 값과 데이터 타입 확인


#train
print("최소 :", x_train[0].min(), "최대 :", x_train[0].max())
print("dtype :", x_train[0].dtype)

#test
print("최소 :", x_test[0].min(), "최대 :", x_test[0].max())
print("dtype :", x_test[0].dtype)


# 첫 번째 이미지 시각화
#train
ax1 = plt.subplot(2, 1, 1)
plt.imshow(x_train[0], cmap = 'gray')
plt.ylabel('train')

#test
ax2 = plt.subplot(2, 1, 2, sharex = ax1)
plt.imshow(x_test[0], cmap = 'gray')
plt.ylabel('test')
```
